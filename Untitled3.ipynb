```python
import pandas as pd

# Load the dataset into a DataFrame
football_df = pd.read_csv("C:/Users/aneli/Downloads/dataset - 2020-09-24 (1).csv")
# Display the first few rows of the dataset
print(football_df.head())

# Get information about the dataset
print(football_df.info())

# Summary statistics
print(football_df.describe())

# Handle missing values (if any)
football_df.dropna(inplace=True)

# Handle duplicates (if any)
football_df.drop_duplicates(inplace=True)


import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(8, 6))
sns.histplot(football_df['Goals'], bins=20, kde=True)
plt.title('Distribution of Goals Scored')
plt.xlabel('Goals')
plt.ylabel('Frequency')
plt.show()

# pScatter plot of goals vs appearances
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Appearances', y='Goals', data=football_df)
plt.title('Goals vs Appearances')
plt.xlabel('Appearances')
plt.ylabel('Goals')
plt.show()

# Example: Box plot of goals by nationality
plt.figure(figsize=(12, 8))
sns.boxplot(x='Nationality', y='Goals', data=football_df)
plt.title('Goals by Nationality')
plt.xlabel('Nationality')
plt.ylabel('Goals')
plt.xticks(rotation=90)
plt.show()

```
